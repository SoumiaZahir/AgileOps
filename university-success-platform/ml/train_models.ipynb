```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score
import mlflow
import optuna

mlflow.set_experiment("All_Models_ExperimentV")
mlflow.set_tracking_uri("http://mlflow:5000")
```

```python

def objective(trial):
    # Choose classifier
    classifier_name = trial.suggest_categorical(
        "classifier", ["LogisticRegression", "RandomForest", "XGBClassifier", "LGBMClassifier"]
    )

    # Define parameters based on selection
    if classifier_name == "LogisticRegression":
        c_val = trial.suggest_float("lr_c", 1e-3, 10, log=True)
        clf = LogisticRegression(C=c_val, random_state=42, max_iter=1000)

    elif classifier_name == "RandomForest":
        n_estimators = trial.suggest_int("rf_n_estimators", 50, 300)
        max_depth = trial.suggest_int("rf_max_depth", 3, 15)
        clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)

    elif classifier_name == "XGBClassifier":
        n_estimators = trial.suggest_int("xgb_n_estimators", 50, 300)
        max_depth = trial.suggest_int("xgb_max_depth", 3, 10)
        learning_rate = trial.suggest_float("xgb_lr", 1e-3, 0.3, log=True)
        clf = XGBClassifier(n_estimators=n_estimators, max_depth=max_depth, learning_rate=learning_rate, random_state=42)

    else: # LGBMClassifier
        n_estimators = trial.suggest_int("lgbm_n_estimators", 50, 300)
        num_leaves = trial.suggest_int("lgbm_num_leaves", 20, 150)
        learning_rate = trial.suggest_float("lgbm_lr", 1e-3, 0.3, log=True)
        clf = LGBMClassifier(n_estimators=n_estimators, num_leaves=num_leaves, learning_rate=learning_rate, random_state=42, verbose=-1)

    # Create a pipeline to handle NaNs automatically
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),
        ('classifier', clf)
    ])

    # Cross-validation on training data
    score = cross_val_score(pipeline, x_train, y_train, cv=3, scoring='accuracy').mean()

    # Log to MLflow
    with mlflow.start_run(run_name=f"Trial_{trial.number}_{classifier_name}", nested=True):
        mlflow.log_param("model_type", classifier_name)
        mlflow.log_params(trial.params)
        mlflow.log_metric("accuracy", score)

    return score

# Run the optimization
study = optuna.create_study(direction="maximize")
with mlflow.start_run(run_name="Main_ML_Search"):
    study.optimize(objective, n_trials=20)

print("Best Trial:", study.best_trial.params)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# 1. طباعة أفضل المعاملات
best_params = study.best_trial.params
print("Best Params:", best_params)

# 2. استخراج اسم أفضل نموذج عثرت عليه الأداة
best_classifier_name = best_params["classifier"]

# 3. بناء النموذج الفائز بناءً على اسمه ومعاملاته
if best_classifier_name == "LogisticRegression":
    clf = LogisticRegression(C=best_params["lr_c"], random_state=42, max_iter=1000)

elif best_classifier_name == "RandomForest":
    clf = RandomForestClassifier(n_estimators=best_params["rf_n_estimators"], max_depth=best_params["rf_max_depth"], random_state=42)

elif best_classifier_name == "XGBClassifier":
    clf = XGBClassifier(n_estimators=best_params["xgb_n_estimators"], max_depth=best_params["xgb_max_depth"], learning_rate=best_params["xgb_lr"], random_state=42)

else: # LGBMClassifier
    clf = LGBMClassifier(n_estimators=best_params["lgbm_n_estimators"], num_leaves=best_params["lgbm_num_leaves"], learning_rate=best_params["lgbm_lr"], random_state=42, verbose=-1)

# 4. إنشاء الـ Pipeline النهائي وتدريبه (Fit)
best_model = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('classifier', clf)
])

# تدريب النموذج الفائز على كامل البيانات
best_model.fit(x_train, y_train)

print("Best model is ready to use!")
# الآن يمكنك استخدام النموذج الفائز للتنبؤ مباشرة: best_model.predict(x_test)
```
